```python
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
file_path = 'ERMI_dataset.csv'
df = pd.read_csv(file_path)
df.head()
```

```python
# df.columns
```

```python
#every column dataframe
#some columns has * so we are removing that using below function 
df.columns = df.columns.str.replace('*', '')
```

```python
len(df.columns)
```

```python
#constat columsn
print(df.columns[df.nunique() == 1])
```

```python
#dropping the constant columns
df = df.loc[:, df.nunique() != 1]
```

```python
len(df.columns)
```

```python
#dropping all empty columsn
df = df.dropna(axis=1, how='all')
```

```python
df = df.drop('Zip',axis=1)
```

```python
len(df.columns)
```

```python
# function to remove asterisks from each cells
def remove_asterisks(s):
  if not isinstance(s, float) and '*' in s:
    return s.replace("*", "")
  return s

# Define the function to convert non-numeric values to 0
def convert_to_float_or_zero(value):
    try:
        return float(value)
    except ValueError:
        return 0
```

```python
# Determine the index of the "ERMI Score (Group I - Group II)" column
mold_start_index = df.columns.get_loc("Aspergillus flavus/oryzae")
mold_end_index = df.columns.get_loc("ERMI Score (Group I - Group II)")
print(f" mold start index {mold_start_index} and end index {mold_end_index}")
```

```python
for column_name in df.columns[mold_start_index:mold_end_index+1]:
    df[column_name] = df[column_name].apply(remove_asterisks)
    df[column_name] = df[column_name].apply(convert_to_float_or_zero)
    df[column_name] = df[column_name].fillna(0)
```

```python
df.head()
```

```python
columns_to_scale = df.columns[mold_start_index:mold_end_index+1]
```

```python
scaler = StandardScaler()
```

```python
df[columns_to_scale] = scaler.fit_transform(df[columns_to_scale])
```

```python
df
```

```python
symp_start_index = df.columns.get_loc("Eyes, Ears, Nose, & Throat")
symp_end_index = df.columns.get_loc("Reproductive")
print(f" Sumptoms start index {symp_start_index} and end index {symp_end_index}")
```

```python
def create_symptom_columns(row):
    for col in range(symp_start_index, symp_end_index + 1):
        symptoms = row[col]
        # Checking for NaN values and if the value is a string before splitting
        if isinstance(symptoms, str) and not pd.isna(symptoms):
            symptoms = symptoms.split(', ')
            for symptom in symptoms:
                df.at[row.name, f"{symptom}"] = 1
```

```python
df.apply(create_symptom_columns, axis=1)
```

```python
diag_index = df.columns.get_loc("Diagnoses")
print(f" Diagnoses index {diag_index}")
```

```python
def create_diagnosis_columns(row):
    diagnoses = row[diag_index]
    if isinstance(diagnoses, str) and not pd.isna(diagnoses):
        diagnoses = diagnoses.split(', ')
        for diagnose in diagnoses:
            df.at[row.name, f"{diagnose}"] = 1
```

```python
df.apply(create_diagnosis_columns, axis=1)
```

```python
#dropping orginal column and resetting the index
diag_symp_to_drop = df.columns[diag_index:symp_end_index + 1]
diag_symp_to_drop
```

```python
df = df.drop(columns=diag_symp_to_drop)
```

```python
repeated_values_to_drop = ['mouth', 'genitals', 'skin', 'internal)']
```

```python
df = df.drop(columns=repeated_values_to_drop)
```
